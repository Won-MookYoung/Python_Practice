```python
## 책7
test=df['최종주문일'][0]   #그냥 되나안되나 해보려고 그중 하나를 뽑아서 테스트한거 
# test[-4:]
test.split(' ') # 공백을 기준으로 나눔

test.split(' ')[-1].split(':')[0]
```

```python
## apply 이용해서 시간 추출

def extract_time(data):
    return data.split(' ')[-1].split(':')[0]
extract_time(test)

df['주문시간']=df['최종주문일'].apply(extract_time)
df['주문시간']
```

```python
## 람다이용
df['주문시간']=df['최종주문일'].apply(lambda x: x.split(' ')[-1])
df['주문시간']
```

```python
#7-2 피벗테이블로 표현하기
#이렇게 하면 인덱스가 1~10 순서로 안나오고 1 10 11 12 2 3 4 이런 순으로 나옴 그래서 int형으로 변환시켜주기

pivot_data=pd.pivot_table(data=df, index='주문시간', values='총판매액',aggfunc=sum).sort_index(ascending=False)
```

```python
# int로 변환후 정렬 
pivot_data.index = pivot_data.index.astype('int') # index의 데이터 타입을 int 로 수정
pivot_data = pivot_data.sort_index()
```
